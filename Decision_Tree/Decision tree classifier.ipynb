```python
### Data import
import pandas as pd

data_uncleaned = pd.read_csv('Train.csv')
data_uncleaned.head(n=10)
```

```python
### Data preprocessing
from sklearn.preprocessing import LabelEncoder ##Convert data to numeric
le = LabelEncoder()

### Remove unwanted columns
columnsToDrop = ['name','ticket','fare','cabin','body','home.dest','embarked','boat']
data_wanted = data_uncleaned.drop(columnsToDrop,axis=1)
data_wanted.head(n=10)
```

```python
data_wanted.info()
print("\n\n\n")
###Fill the missing data
data_filled = data_wanted.fillna(data_wanted['age'].mean())
data_filled.info()
```

```python
data_filled['sex'] = le.fit_transform(data_filled["sex"])
data_filled.head(n=10)
```

```python
input_cols = ['pclass',"sex","age","sibsp","parch"]
output_cols = ["survived"]

X = data_filled[input_cols]
Y = data_filled[output_cols]

print(X.shape,Y.shape)
print(type(X))
```

```python
### Model
"""Written by deepanshu garg"""

import numpy as np
def entropy(col):
    """Returns the entropy of the col or Y"""
    ent = 0.0
    counts = np.unique(col,return_counts = True)
    for i in range(len(counts[0])):
        p = counts[1][i]/sum(counts[1])
        ent += -1.0*p*np.log2(p) 
    return ent

def divide_data(data,fval,fcol):
    """Divide the data into two parts"""
    x_right = pd.DataFrame([],columns=data.columns)
    x_left = pd.DataFrame([],columns=data.columns)
    for ix in range(data.shape[0]):
        val = data[fcol].loc[ix]
        if val > fval:
            x_right = x_right.append(data.loc[ix])
        else:
            x_left = x_left.append(data.loc[ix])
    return x_left,x_right

def information_gain(data,fval,fcol):
    """Function return the information gain wrt entropy for the particular col"""
    
    data_left,data_right = divide_data(data,fval,fcol)
    
    l = data_left.shape[0]/data.shape[0]
    r = data_right.shape[0]/data.shape[0]
    
    if data_left.shape[0] == 0 or data_right.shape[0] ==0:
        return -1000000 #Min Information Gain
    
    ig = entropy(data.survived) - (l*entropy(data_left.survived) + r*entropy(data_right.survived))
    return ig
```

```python
for fx in X.columns:
    print(fx)
    print(information_gain(data_filled,data_filled[fx].mean(),fx))
```

```python
class DecisionTree:
    def __init__(self,depth=0,max_depth=5):
        self.left = None
        self.right = None
        self.fkey = None
        self.fval = None
        self.max_depth = max_depth
        self.depth = depth
        self.target = None
        
    def train(self,X_train):
        
        features = ['pclass','sex','age','sibsp', 'parch']
        info_gains = []
        
        for ix in features:
            i_gain = information_gain(X_train,X_train[ix].mean(),ix)
            info_gains.append(i_gain)
            
        self.fkey = features[np.argmax(info_gains)]
        self.fval = X_train[self.fkey].mean()
        print("Making Tree Features is",self.fkey)
        
        #Split Data
        data_left,data_right = divide_data(X_train,self.fval,self.fkey)
        data_left = data_left.reset_index(drop=True)
        data_right = data_right.reset_index(drop=True)
         
        #Truly a left node
        if data_left.shape[0]  == 0 or data_right.shape[0] ==0:
            if X_train.survived.mean() >= 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
        #Stop earyly when depth >=max depth
        if(self.depth>=self.max_depth):
            if X_train.survived.mean() >= 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
        
        #Recursive Case
        self.left = DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.left.train(data_left)
        
        self.right = DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.right.train(data_right)
        
        #You can set the target at every node
        if X_train.survived.mean() >= 0.5:
            self.target = "Survive"
        else:
            self.target = "Dead"
        return
    
    def predict(self,test):
        if test[self.fkey]>self.fval:
            #go to right
            if self.right is None:
                return self.target
            return self.right.predict(test)
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)
```

```python
split = int(0.7*data_filled.shape[0])
train_data = data_filled[:split]
test_data = data_filled[split:]
test_data = test_data.reset_index(drop=True)
```
